User column contains all user-related data in the form of the dictionary. For the reserach purposes username (unique field) will be sufficient out all of it. Create a separate dataframe with user details:

```python
def get_users(df, name):
    keys = list(df[name])[0].keys()
    user_data = pd.DataFrame(index = df.index, columns = keys)
    for ind in df.index:
        for key in keys:
            buf = df.loc[ind, name][key]
            user_data.loc[ind, key] = str(buf)
    return user_data
```

```python
users_qatar = get_users(qatar_tweets, "user")
users_brazil = get_users(brazil_tweets, "user")
users_russia = get_users(russia_tweets, "user")
```

**Flesch Reading Ease and Flesch Kincaid Grade coefficients:**

"The **FleschKincaid** readability tests are readability tests designed to indicate how difficult a passage in English is to understand. There are two tests: the **Flesch Reading-Ease**, and the **FleschKincaid Grade Level**. Although they use the same core measures (word length and sentence length), they have different weighting factors.

The results of the two tests correlate approximately inversely: a text with a comparatively high score on the Reading Ease test should have a lower score on the Grade-Level test. Rudolf Flesch devised the Reading Ease evaluation; somewhat later, he and J. Peter Kincaid developed the Grade Level evaluation for the United States Navy." [1]

**`Step 6.` Lower case**

**`Step 7.` Remove stop words, punctuation, non-unicode charachters, links, e-mails, @tagging**

Stop words are the most commonly used words in the given language, which thus carry no specific information about the subject. In order to perform data analysis on texts we need to extract from them features which captures specific the best.

```python
# define stopwords
stopwords_list = list(stopwords.words('english'))
```

Any kind on non-alpha charachters are not useful for NLP either, since they actually bears no precious infromation aboyut the subject and can be used with no regard to sentiment, topic, style, etc.

```python
# defineregex patterms 
non_unicode = r"[^\x00-\x7F]+\ *(?:[^\x00-\x7F]| )*"
url = r"(https?:\/\/)?(www\.)?[-a-zA-Z0-9@:%._\+~#=]{1,256}\.[a-zA-Z0-9()]{1,6}\b([-a-zA-Z0-9()@:%_\+.~#?&//=]*)" 
email = r"[a-z0-9!#$%&'*+/=?^_`{|}~-]+(?:\.[a-z0-9!#$%&'*+/=?^_`{|}~-]+)*@(?:[a-z0-9](?:[a-z0-9-]*[a-z0-9])?\.)+[a-z0-9](?:[a-z0-9-]*[a-z0-9])?"
tag =  r"(^|\W)@\b([-a-zA-Z0-9._]{3,25})\b"
punctuation = r"\W+"

regex_patterns = [non_unicode, url, email, tag, punctuation]
```

**source**: https://uibakery.io/regex-library/url , https://regex101.com/r/wZo2wB/1

Helping function to remove stop words, punctuation, non-unicode charachters, links, e-mails, @tagging

```python
def clean_text(df, column, stopwords_list, regex_patterns):
    
    #stopwords
    df[column] = df[column].apply(lambda x: " ".join([word for word in str(x).split() if word not in stopwords_list and word.isalpha()]))
    
    #regex
    for pattern in regex_patterns:
        df[column] = df[column].apply(lambda x: re.sub(pattern, " ", x))
```

```python
clean_text(qatar_tweets, "content", stopwords_list, regex_patterns)
clean_text(russia_tweets, "content", stopwords_list, regex_patterns)
clean_text(brazil_tweets, "content", stopwords_list, regex_patterns)
```

**`Step 1.` Collocations**

In simple terms, collocations are phrases in whic words are connected semanticly. There might be 2, 3 or more quantity of words following restricted sequence. For this research I am findning bigrams - collocations containing 2 tokens and visualize frequency of use of each. This excercise captures the essence of the given corpora - in this case subset of tweets for each country, given FIFA and time context.



```python
def find_bigrams (df, name, country, n):
    df_text = " ".join(df[name])
    df_text_bigrams = pd.DataFrame((pd.Series(nltk.bigrams(nltk.word_tokenize(df_text))).value_counts()), columns = ["bigrams"])
    df_text_bigrams["country"] = country
    
    try:
        return df_text_bigrams.head(n)
    except:
        return df_text_bigrams
```

**`Step 3.` Sentiment analysis**

Sentiment analysis carrying out witht he help of nltk.SentimentIntensityAnalyzer. It returns the dictionary with calculated probabilities for positive, negative and neutral sentiment. Compound component is computed by normalizing all three scores. 

```python
sia = SentimentIntensityAnalyzer()
sia.polarity_scores("Wow, NLTK is really powerful!")

tweets_together = [qatar_tweets, brazil_tweets, russia_tweets]
for cnt in tweets_together:
    cnt['sentiment'] = cnt['content'].apply(lambda x: sia.polarity_scores(x))
   
```

**Source:** https://towardsdatascience.com/sentimental-analysis-using-vader-a3415fef7664

```python
qatar_tweets.head(3)
```

Now I need parse sentiment from the dictionary:

```python
for cnt in tweets_together:
    max_value = 0
    for ind in cnt.index:
        dict_ = cnt.loc[ind, 'sentiment']
        max_value = 0
        for key in dict_:
            if key != "compound" and dict_ [key] > max_value:
                max_value = dict_[key]
                cnt.loc[ind, 'sentiment'] = str(key)
```
