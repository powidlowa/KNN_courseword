```python
def plot_pie(languages_selected):
  fig, ax = plt.subplots(figsize=(10, 10)) #size of the plot
  size = 0.7 #size of the white inner circle

  # equally distanced one from another shades of pallette 'viridis'
  colours = plt.get_cmap('rocket')(np.linspace(0.1, 0.6, len(languages_selected.index)))

  # pie chart itself, returns properties
  wedges, texts = ax.pie(languages_selected["count"], 
                        colors = colours,
                        startangle = 0,
                        rotatelabels = True,
                        wedgeprops = dict(width = size, edgecolor = 'white', linewidth = 1))

  # setting properties for boxes for annotations (aka lables)
  bbox_props = dict(boxstyle="round, pad = 0.3", fc = "white", lw = 0.6)

  # setting properties for arrows linking boxes and pie chart itself
  kw = dict(arrowprops = dict(arrowstyle="-["),
            bbox = bbox_props, va = "center")

  # construction of annotations: text + boxes + arrows
  for i, p in enumerate(wedges):
      ang = (p.theta2 - p.theta1)/2. + p.theta1
      y = np.sin(np.deg2rad(ang))
      x = np.cos(np.deg2rad(ang))
      horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
      connectionstyle = "angle, angleA = 0, angleB = {}".format(ang)
      kw["arrowprops"].update({"connectionstyle": connectionstyle})
      ax.annotate(str(languages_selected.index[i] + " :\n" + str(languages_selected["count"][i]) + " tweets or ~" + \
                  str(round(languages_selected["count"][i]*100/sum(languages_selected["count"]))) + "%"), 
                  xy=(x, y), xytext=(1.1*np.sign(x), 1.4*y),
                  horizontalalignment=horizontalalignment, **kw)

  ax.set(aspect="equal")
  plt.title('\nLANGUAGE USAGE\n', fontsize = 18, loc = 'center')
  plt.rcParams['font.size'] = 15
  plt.show()
```

`Step 1.` Text pre-processing steps to ensure data is suitable for Neural Network Processing:
- lower case
- remove tokens

- pivoting datasets if needed to the format label-sample tabular form.

```python
def clean_text(df, column, stopwords_list, regex_patterns, lemmatizer):
    
    #stopwords removal
    df[column] = df[column].apply(lambda x: " ".join(
                                  [word for word in str(x).split() if word not in stopwords_list \
                                  and word.isalpha()]))
    #regex
    for pattern in regex_patterns:
        df[column] = df[column].apply(lambda x: re.sub(pattern, " ", x))
        
    #lemmatize   
    df[column].apply(lambda x: "".join([lemmatizer.lemmatize(w) for w in x]))
    
    #lower case
    df[column] = df[column].str.lower()
    
    #add column for tokens count
    df["tokens"] = df[column].apply(lambda x: int(len(nltk.word_tokenize(x))))
    

    return df
```

```python
# define stopwords
stopwords_list = list(stopwords.words('english'))

# defineregex patterms 
non_unicode = r"[^\x00-\x7F]+\ *(?:[^\x00-\x7F]| )*"
url = r"(https?:\/\/)?(www\.)?[-a-zA-Z0-9@:%._\+~#=]{1,256}\.[a-zA-Z0-9()]{1,6}\b([-a-zA-Z0-9()@:%_\+.~#?&//=]*)" 
email = r"[a-z0-9!#$%&'*+/=?^_`{|}~-]+(?:\.[a-z0-9!#$%&'*+/=?^_`{|}~-]+)*@(?:[a-z0-9](?:[a-z0-9-]*[a-z0-9])?\.)+[a-z0-9](?:[a-z0-9-]*[a-z0-9])?"
tag =  r"(^|\W)@\b([-a-zA-Z0-9._]{3,25})\b"
punctuation = r"\W+"

regex_patterns = [non_unicode, url, email, tag, punctuation]

# initialize Lemmatizer from nltk
wnl = WordNetLemmatizer()
```

After excluding top and bottom 10% of values mean and median became closer to each other which indicates dataset became Since there is no possibility to actually oversample dataset, I will udnersample emotions which to the roughly mean +10% value after exlusion of outliers.

```python
def updersample (df, max_value, criteria_column, sort_by):
  
  criteria_unique = list(np.unique(df[criteria_column]))
  df_balanced = pd.DataFrame()
  count = 0
  # loop through separate dataframes for each criteria
  for criteria in criteria_unique:
    buffer_df = df[df[criteria_column] == criteria]
    # first remove short records
    buffer_df = buffer_df.sort_values(by = sort_by, ascending = False)

    if buffer_df.shape[0] > max_value:
      if count == 0:
        df_balanced = buffer_df.iloc[:max_value,:].copy()
      else:
        df_balanced = pd.concat([df_balanced, 
                                 buffer_df.iloc[:max_value,:]])

    elif buffer_df.shape[0] < max_value:  
      if count == 0:
        df_balanced = buffer_df.iloc[:max_value,:].copy()
      else:
        df_balanced = pd.concat([df_balanced, 
                                 buffer_df.iloc[:max_value,:]])

    count = count + 1

  return df_balanced
```

```python
class BERT_assistant (object):
    
    """
    Initialization method
    """
    def __init__(self, tweets, emotions_unique, 
                 emotions_full, preprocess_layer, encoder_layer, ):
     
        # initialize X and Y
        self.X = tweets
        self.y = None
        
        # initialize train and test subsetst for each
        self.X_train = None
        self.X_test = None 
        self.y_train = None
        self.y_test = None
        
        # field to store prediction
        self.y_pred = []
        # field to store prediction probabilities of each class
        self.y_pred_probs = None
        # field to store preprocess_layer
        self.preprocessing_layer = preprocess_layer
        # field to store encoder_layer
        self.encoder_layer = encoder_layer
        # field to store model
        self.model = None
        # lists of emotions (= class labels) - unique and full
        self.emotions_unique = emotions_unique
        self.emotions_full = emotions_full
        
        # Dictionary with encoded emotions if the form {emotions:index}, 
        # where index = 0,1 or 2
        self.encode_dict = {k:v for k,v in zip(self.emotions_unique, 
                                        np.arange(len(self.emotions_unique)))}
        # decode dictionary
        self.decode_dict = {k:v for k,v in zip(np.arange(len(self.emotions_unique)),
                                               self.emotions_unique)}

        # Initialize field for keras history object
        self_history = None
       
    """
    Encode labels and samples; split data and lables into trainit 
    and test subsets
    """
    def encode_n_split (self):

        # encode lables
        self.y = np.array([self.encode_dict[emo] for emo in self.emotions_full])
        # split training set
        self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(
                                                     self.X, self.y, 
                                                     test_size=0.2, 
                                                     random_state=42)
    """
    Model initialization and bulding (adding layers)
    """
    # build model
    def build_model(self, num_classes, dropout):
        
        
        # input layer
        text_input = tf.keras.layers.Input(shape=(), dtype=tf.string, 
                                           name='text')
        # preprocessing layer BERT -> padding is done here
        preprocessing_layer = hub.KerasLayer(self.preprocessing_layer, 
                                             name='preprocessing')
        # encoding layer BERT
        encoder_inputs = preprocessing_layer( text_input)
        encoder = hub.KerasLayer(self.encoder_layer, trainable=True, 
                                 name='BERT_encoder')
        outputs = encoder(encoder_inputs)
        net = outputs['pooled_output']
        # Dropout layer
        net = tf.keras.layers.Dropout(dropout)(net)
        # Output layer
        net = tf.keras.layers.Dense(num_classes, 
                                    activation='sigmoid', # probs between 0 and 1
                                    name='classifier')(net)
        self.model = tf.keras.Model(text_input, net)
          
    """
    Print out model architecture
    """
    
    def print_architecture(self):
      tf.keras.utils.plot_model(self.model)

    """
    Compile and fit model onto data
    """
    #add patience as parametr
    def fit_mod(self, epochs, history_flag = False, patience = 1): 

        self.model.compile(optimizer = 'adamax', 
                           loss = 'sparse_categorical_crossentropy', 
                           metrics = ["acc"])
        if history_flag:
            self.history = self.model.fit(self.X_train, self.y_train,
                      epochs = epochs,
                      batch_size = 128,
                      validation_split = 0.2,
                      callbacks = [EarlyStopping(monitor = 'val_loss', 
                                                 patience = patience)])        
            return self.history    
        else:
            self.model.fit(self.X_train, self.y_train,
                           epochs = epochs,
                           batch_size = 128,
                           validation_split = 0.2,
                           callbacks = [EarlyStopping(monitor = 'val_loss', 
                                                      patience = patience)])
    
    """
    Predict
    """
    def predict(self, dataset):

        self.y_pred = []
        self.y_pred_probs = self.model.predict(dataset) #e.g. self.X_test 
        
        for probs in self.y_pred_probs:
            self.y_pred.append(np.argmax(probs))
        
        return self.y_pred
```

```python
# confusion matrix sent
def confusion_matr_percentage(y_test, y_pred, num_classes, name, classes):
  conf_matrix = tf.math.confusion_matrix(y_test, y_pred, 
                                         num_classes=num_classes, 
                                         weights=None, 
                                         dtype=tf.dtypes.int32, name=None)

  line_sum = 0
  new_conf_matrix = pd.DataFrame(data = np.zeros((num_classes,num_classes)),
                                 index = classes,
                                 columns = classes)

  for i in range(num_classes):
      line_sum = int(sum(conf_matrix[i]))
      for j in range(num_classes):
        try:
          new_conf_matrix.iloc[i,j] = np.round(int(conf_matrix[i,j])/line_sum, 2)
        except:
          new_conf_matrix.iloc[i,j] = 0

        
  fig, ax = plt.subplots(figsize=(10,8))
  print("\n\nConfusion matrix %s in per-cent: \n"%name)
  sns.heatmap(new_conf_matrix, annot = new_conf_matrix, fmt = '.2g')
```
