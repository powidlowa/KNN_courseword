```python
def plot_pie(dataframe, groupby_column, select_column, angle, title, 
             size=(7, 7), groupby=True, cmap='rocket', ax="null"):
    
    if groupby:
        result_df = pd.DataFrame(
            dataframe
            .groupby(groupby_column)
            .count()[select_column]
        )
    else:
        result_df = dataframe
    
    if ax=="null":
        fig, ax = plt.subplots(figsize=size) #size of the plot
    size = 0.7 #size of the white inner circle

    # equally distanced one from another shades of pallette 'viridis'
    colours = plt.get_cmap(cmap)(np.linspace(0.1, 0.6, result_df.shape[0]))

    # pie chart itself, returns properties
    wedges, texts = ax.pie(result_df[select_column], 
                            colors = colours,
                            startangle = angle,
                            rotatelabels = True,
                            wedgeprops = dict(width = size, edgecolor = 'white', linewidth = 1))

    # setting properties for boxes for annotations (aka lables)
    bbox_props = dict(boxstyle="round, pad = 0.3", fc = "white", lw = 0.2)

    # setting properties for arrows linking boxes and pie chart itself
    kw = dict(arrowprops = dict(arrowstyle="-["),
                bbox = bbox_props, va = "center")

    # construction of annotations: text + boxes + arrows
    for i, p in enumerate(wedges):
        ang = (p.theta2 - p.theta1)/2. + p.theta1
        y = np.sin(np.deg2rad(ang))
        x = np.cos(np.deg2rad(ang))
        horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
        connectionstyle = "angle, angleA = 0, angleB = {}".format(ang)
        kw["arrowprops"].update({"connectionstyle": connectionstyle})
        ax.annotate(str(result_df.index[i] + " :\n" 
                    + str(result_df[select_column][i]) + " responses or ~" 
                    + str(round(result_df[select_column][i]*100/sum(result_df[select_column]))) 
                    + "%"), 
                    xy=(x, y), xytext=(1.1*np.sign(x), 1.1*y),
                    horizontalalignment=horizontalalignment, **kw)

    ax.set(aspect="equal")
    plt.title(title, fontsize = 14, loc = 'center')
    plt.rcParams['font.size'] = 15
    
    sns.reset_orig()
    plt.show()
```

```python
def plot_barplot(dataframe, groupby_column, select_column, orient, 
                 title, axes_i, size=(9, 7), groupby=True, label_type='edge', cmap='rocket'):
    
    if groupby:
        result_df = pd.DataFrame(
            dataframe
            .groupby(groupby_column)
            .count()[select_column]
        )
        percentage = list(np.round(result_df[select_column]/int(result_df.sum())*100, 2))
        lables_perc = [str(p)+"%" for p in percentage]
    else:
        result_df = dataframe

    # set style/theme
    colours = plt.get_cmap(cmap)(np.linspace(0.1, 0.6, result_df.shape[0]))
    custom_params = {"axes.spines.right": False, "axes.spines.top": False, 'figure.figsize':size}
    sns.set_theme(style="ticks", rc=custom_params)

    # build plot and adjust
    if axes_i=="null":
        bp = sns.barplot(data=result_df.T, orient=orient, palette=colours)
        bp.set_title(title, fontsize=14)
        bp.set(ylabel=None)
        if groupby:
            bp.bar_label(bp.containers[-1], label_type=label_type, labels=lables_perc)
        else:
            bp.bar_label(bp.containers[-1], label_type=label_type)
    else:
        sns.barplot(ax=axes_i, data=result_df.T, orient=orient, palette=colours)
        axes_i.set_title(title, fontsize=18)
        axes_i.set(ylabel=None)
        if groupby:
            axes_i.bar_label(axes_i.containers[-1], label_type=label_type, labels=lables_perc)
        else:
            axes_i.bar_label(axes_i.containers[-1], label_type=label_type)

    
    sns.reset_orig()
```

```python
def plot_geodata(df, user_agent, country_column, id_column):
    
    # enrich dataframe with geo data
    geolocator=Nominatim(user_agent=user_agent)
    
    def geolocate(country, ret):
        try:
            # Geolocate the center of the country
            loc = geolocator.geocode(country)

            if ret=="latitude":
                return loc.latitude
            if ret=="longitude":
                return loc.longitude
            else:
                return (loc.latitude, loc.longitude)
        except:
            # Return missing value
            return np.nan
    
    
    df["latitude"] = (
        df[country_column].apply(lambda x: geolocate(x, "latitude"))
    )
    df["longitude"] = (
        df[country_column].apply(lambda x: geolocate(x, "longitude"))
    )

    # create empty map
    world_map= folium.Map(tiles="cartodbpositron")
    marker_cluster = MarkerCluster().add_to(world_map)

    # for each coordinate, create circlemarker of user percent
    for i in range(len(df)):
        lat = df.iloc[i]['latitude']
        long = df.iloc[i]['longitude']
        radius=5
        popup_text = """Country : {}<br>
                        %of Users : {}<br>"""
        popup_text = popup_text.format(df.iloc[i][country_column],
                                           df.iloc[i][id_column]
            )
        folium.CircleMarker(location = [lat, long], 
                                radius=radius, 
                                popup=popup_text, 
                                fill=True).add_to(marker_cluster)
    # show the map
    return world_map
```

```python
def text_wrap(list_to_wrap, length):
    
    wraped_text=[]
    wraped_text_dict={}

    for text in list_to_wrap:
        wraped_text.append(
            "\n".join(
                textwrap.wrap(text, length)
            )
        )
    
    renaming_dict={k:v for k,v in zip(list_to_wrap, wraped_text)}
    return renaming_dict
```

```python
%%js

// Run this cell to update your word count.

function wordcount() {
    let wordCount = 0
    let extraCount = 0
    let mainBody = true

    let cells = Jupyter.notebook.get_cells()
    cells.forEach((cell) => {
        if (cell.cell_type == 'markdown') {
            let text = cell.get_text()
            // Stop counting as main body when get to References or Appendices.
            if (text.startsWith('## References') ||
                text.startsWith('## Appendices')) {
                mainBody = false
            }
            if (text.startsWith('## Word Count')) {
                text = ''
            }
            if (text) {
                let words = text.toLowerCase().match(/\b[a-z\d]+\b/g)
                if (words) {
                    let cellCount = words.length
                    if (mainBody) {
                        wordCount += cellCount
                    } else {
                        extraCount += cellCount
                    }
                }
            }
        }
    })
    return [wordCount, extraCount]
}

let wc = wordcount()
element.append(`Main word count: ${wc[0]} (References and appendices word count: ${wc[1]})`)
```
